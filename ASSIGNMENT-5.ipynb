Q)1:-Guess a number is a game that prompts a player to guess a number between 0 and 9, which
is randomly generated by the system. When the input given by the user matches the
number generated by the system then the user wins. The game should go as follows:

```python
import random 
num=random.randint(0,9)
while num==num:
    gusses=eval(input('enter the number:-'))
    if num==gusses:
        print('you got it right!congo!')
        break
    else:
        print('sorry try again')
```
